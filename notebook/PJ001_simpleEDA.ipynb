```python
#ファイルの読込
df = pd.read_excel('../csv/内科カンファ数値.xlsx', index_col=0)
```

```python
# データの転置とインデックスのリセット
df_transposed = df.T.reset_index()
df_transposed = df_transposed.rename(columns={'index': '日付'})
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='日付', y='Plt', data=df_transposed, marker='o', label='Plt')
sns.lineplot(x='日付', y='CRP', data=df_transposed, marker='o', label='CRP')

plt.xlabel('日付', fontsize=14)
plt.ylabel('値', fontsize=14)
plt.title('PltとCRPの推移', fontsize=20)
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
